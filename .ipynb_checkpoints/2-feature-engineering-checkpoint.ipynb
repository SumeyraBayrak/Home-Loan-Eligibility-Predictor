<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>Feature Engineering</h2>

Hedef değişkeni etkileyebilecek yeni özellikler oluşturarak analizimize daha fazla anlam katabiliriz. Aşağıda, oluşturmayı planladığımız üç yeni özellik ve bu özelliklerin dayandığı mantık, önemli İngilizce terimlerle birlikte detaylı bir şekilde açıklanmıştır:

<h4>1. Toplam Gelir (<strong>Total Income</strong>):</h4>  
Başvuru sahibinin geliri (*Applicant Income*) ile eşin geliri (*Coapplicant Income*) birleştirilerek toplam gelir değeri oluşturulacaktır. Bu özellik, <strong>financial strength</strong> (finansal güç) ve gelir düzeyinin (*<strong>income level</strong>*) kredi onaylama sürecindeki rolünü vurgulamaktadır. Yüksek toplam gelir, bireyin kredi ödeme kapasitesinin arttığını ve kredi onaylanma olasılığının yükseldiğini gösterir. Bu, <strong>creditworthiness</strong> (kredi değerliliği) üzerinde doğrudan etkili olabilir.

<h4>2. EMI (<strong>Equated Monthly Installment</strong>):</h4>  
EMI, başvuru sahibinin krediyi geri ödemek için her ay düzenli olarak ödemesi gereken sabit taksit tutarını ifade eder. Bu, kredi tutarının (*Loan Amount*) kredi vadesine (*Loan Term*) bölünmesiyle hesaplanabilir. Yüksek EMI, başvuru sahibinin <strong>debt burden</strong> (borç yükü) üzerinde baskı oluşturabilir ve bu da geri ödeme güçlüğü (*repayment difficulty*) yaşama olasılığını artırabilir. Bu değişken, başvuru sahibinin ödeme performansını ve <strong>loan approval likelihood</strong> (kredi onaylanma olasılığı) üzerindeki etkilerini anlamada önemli bir rol oynar.

<h4>3. Bakiye Gelir (<strong>Balance Income</strong>):</h4>  
Bakiye gelir, EMI ödendikten sonra başvuru sahibinin elinde kalan gelir miktarını ifade eder. Bu değişken, başvuru sahibinin <strong>disposable income</strong> (harcanabilir gelir) seviyesini ölçer. Yüksek bakiye gelir, kişinin kredi ödemelerini daha rahat karşılayabileceğini ve dolayısıyla kredi geri ödeme kapasitesinin arttığını gösterir. Bu da <strong>loan repayment ability</strong> (kredi geri ödeme yeteneği) üzerinde olumlu bir etkiye sahiptir ve kredi onaylanma ihtimalini yükseltebilir.<br><br>

Bu yeni özellikler, modelin doğruluğunu ve açıklayıcılığını artırarak, <strong>predictive accuracy</strong> (öngörü doğruluğu) ve <strong>risk assessment</strong> (risk değerlendirmesi) süreçlerini geliştirmeyi amaçlamaktadır.
</font>


## Kütüphaneleri İçe Aktarma:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Grafiklerin doğrudan not defterinde gösterilmesini sağlar
%matplotlib inline

# Uyarıları yok say
import warnings
warnings.filterwarnings("ignore")
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>
Veri Setlerinin Yüklenmesi
</h2>
Aşağıdaki kod, iki farklı CSV dosyasını okuyarak eğitim ve test veri setlerini yükler:
</font>

```python
train = pd.read_csv(r"C:\Users\Samira\Sumeyra\btk-akademi\Home-Loan-Eligibility-Predictor\dataset\train_ctrUa4K.csv")
test = pd.read_csv(r"C:\Users\Samira\Sumeyra\btk-akademi\Home-Loan-Eligibility-Predictor\dataset\test_lAUu6dG.csv")
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>
Eğitim Veri Setindeki Eksik Verilerin Doldurulması
</h2>
Aşağıdaki kod, eğitim veri setindeki eksik (NaN) değerleri, her sütunun mod (en sık görülen değer) ile doldurur:
</font>

```python
train["Gender"].fillna(train["Gender"].mode()[0], inplace = True)
train["Married"].fillna(train["Married"].mode()[0], inplace = True)
train["Dependents"].fillna(train["Dependents"].mode()[0], inplace = True)
train["Self_Employed"].fillna(train["Self_Employed"].mode()[0], inplace = True)
train["Credit_History"].fillna(train["Credit_History"].mode()[0], inplace = True)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>
Test Veri Setindeki Eksik Verilerin Doldurulması
</h2>
Aşağıdaki kod, test veri setindeki eksik (NaN) değerleri, her sütunun mod (en sık görülen değer) ile doldurur:
</font>

```python
test["Gender"].fillna(test["Gender"].mode()[0], inplace = True)
test["Married"].fillna(test["Married"].mode()[0], inplace = True)
test["Dependents"].fillna(test["Dependents"].mode()[0], inplace = True)
test["Self_Employed"].fillna(test["Self_Employed"].mode()[0], inplace = True)
test["Credit_History"].fillna(test["Credit_History"].mode()[0], inplace = True)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>
Loan_Amount_Term Sütunundaki Eksik Verilerin Doldurulması
</h2>
Aşağıdaki kod, hem eğitim (<strong>train</strong>) hem de test (<strong>test</strong>) veri setlerinde, <strong>Loan_Amount_Term</strong> sütunundaki eksik (<strong>NaN</strong>) değerleri en sık görülen değer (<strong>mod</strong>) ile doldurur:
</font>

```python
train["Loan_Amount_Term"].fillna(train["Loan_Amount_Term"].mode()[0], inplace = True)
test["Loan_Amount_Term"].fillna(test["Loan_Amount_Term"].mode()[0], inplace = True)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
<h2>
LoanAmount Sütunundaki Eksik Verilerin Doldurulması
</h2>
Aşağıdaki kod, hem eğitim (<strong>train</strong>) hem de test (<strong>test</strong>) veri setlerinde, <strong>LoanAmount</strong> sütunundaki eksik (<strong>NaN</strong>) değerleri sütunun <strong>medyan</strong> değeriyle doldurur:
</font>

```python
train["LoanAmount"].fillna(train["LoanAmount"].median(), inplace = True)
test["LoanAmount"].fillna(test["LoanAmount"].median(), inplace = True)
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">
Total Income'ın dağılımını kontrol edelim:
</font>

```python
# Total Income hesaplama
train["Total_Income"] = train["ApplicantIncome"] + train["CoapplicantIncome"]
test["Total_Income"] = test["ApplicantIncome"] + test["CoapplicantIncome"]

# Yan yana grafikler için subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Eğitim veri setindeki Total Income dağılımı
sns.distplot(train["Total_Income"] , ax=axes[0])
axes[0].set_title("Train Total Income Distribution")

# Test veri setindeki Total Income dağılımı
sns.distplot(test["Total_Income"], ax=axes[1])
axes[1].set_title("Test Total Income Distribution")

plt.tight_layout()
plt.show()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">

Dağılımın sola kaydığını, yani sağa çarpık olduğunu görebiliyoruz. Bu nedenle, dağılımı normal hale getirmek için logaritma dönüşümü uygulayalım:

</font>

```python
# Logarithmic Total Income hesaplama
train["Total_Income_log"] = np.log(train["Total_Income"])
test["Total_Income_log"] = np.log(test["Total_Income"])

# Yan yana grafikler için subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Eğitim veri setindeki Total_Income_log dağılımı
sns.distplot(train["Total_Income_log"], ax=axes[0])
axes[0].set_title("Train Total_Income_log Distribution")

# Test veri setindeki Total_Income_log dağılımı
sns.distplot(test["Total_Income_log"], ax=axes[1])
axes[1].set_title("Test Total_Income_log Distribution")

plt.tight_layout()
plt.show()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">

Şimdi dağılım, normal dağılıma çok daha yakın görünüyor ve uç değerlerin etkisi önemli ölçüde azalmış durumda. Şimdi EMI özelliğini oluşturalım:

</font>


```python
# EMI hesaplama
train["EMI"] = train["LoanAmount"] / train["Loan_Amount_Term"]
test["EMI"] = test["LoanAmount"] / test["Loan_Amount_Term"]

# Yan yana grafikler için subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Eğitim veri setindeki EMI dağılımı
sns.distplot(train["EMI"], ax=axes[0])
axes[0].set_title("Train EMI Distribution")

# Test veri setindeki EMI dağılımı
sns.distplot(test["EMI"], ax=axes[1])
axes[1].set_title("Test EMI Distribution")

plt.tight_layout()
plt.show()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">

Şimdi Balance Income özelliğini oluşturalım ve dağılımını kontrol edelim:

</font>


```python
# Balance Income hesaplama
train["Balance_Income"] = train["Total_Income"] - (train["EMI"] * 1000)
test["Balance_Income"] = test["Total_Income"] - (test["EMI"] * 1000)

# Yan yana grafikler için subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Eğitim veri setindeki Balance Income dağılımı
sns.distplot(train["Balance_Income"], kde=True, ax=axes[0])
axes[0].set_title("Train Balance Income Distribution")

# Test veri setindeki Balance Income dağılımı
sns.distplot(test["Balance_Income"], kde=True, ax=axes[1])
axes[1].set_title("Test Balance Income Distribution")

plt.tight_layout()
plt.show()
```

<font size="4" style="padding: 10px; font-family: Arial, sans-serif; line-height: 1.6; display: block; margin-bottom: 15px; text-align: justify;">

Şimdi, bu yeni özellikleri oluşturmak için kullandığımız eski değişkenleri silelim. Bunun nedeni, yeni eklediğimiz özelliklerle eski özellikler arasındaki yüksek korelasyonun olmasıdır. <strong>Lojistik regresyon</strong>, değişkenler arasındaki yüksek korelasyonun modelin doğruluğunu olumsuz etkileyebileceğini varsayar. Ayrıca, veri setindeki gürültüyü azaltmak için korelasyonu yüksek olan özellikleri kaldırmak, modelin daha sağlam ve doğru sonuçlar vermesine yardımcı olacaktır. 📈🔍

</font>
