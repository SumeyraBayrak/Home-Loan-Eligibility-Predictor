```python
import matplotlib.pyplot as plt
import numpy as np

# Model isimleri
models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']

# Validation ve Leaderboard skorları
validation_scores = [0.8, 0.68, 0.8, 0.78]
leaderboard_scores = [0.81, 0.9, 0.82, 0.87]

# Çubukların konumları
x = np.arange(len(models))
width = 0.35  # Çubukların genişliği

# Grafik oluşturma
plt.figure(figsize=(8, 6))
plt.bar(x - width/2, validation_scores, width, label='Validation score', color='#6C7B8B', edgecolor='black', linewidth=1.2)
plt.bar(x + width/2, leaderboard_scores, width, label='Leaderboard score', color='#9B59B6', edgecolor='black', linewidth=1.2)

# Eksen etiketleri ve başlık
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Validation vs Leaderboard Scores for Models')
plt.xticks(x, models)
plt.ylim(0, 0.99)

# Legend ekleme
plt.legend()

# Grafiği kaydetme (plt.show()'dan önce)
plt.tight_layout()
plt.savefig('validation_vs_leaderboard_scores.png')

# Grafiği gösterme
plt.show()
```
